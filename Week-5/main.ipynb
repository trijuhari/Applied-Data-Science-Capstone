### As we can see in the chart above, the result score is not the best

The explanation of this is that silhouette score range between [-1, +1]. Where values near to +1 indicate that the sample is far away from the neighboring clusters. Values near to 0, this means the clusters are near to the decision boundary, and a score near to -1, is probably a wrong assignation of values to the clusters.

So, to avoid it, I'll use PCA with the motivation of reduce the number of features.

```python
pca_df = PCA().fit(neighbourhoods_groupped_clustering)

plt.rcParams["figure.figsize"] = (22,6)

fig, ax = plt.subplots()
xi = np.arange(1, 96, step=1)
y = np.cumsum(pca_df.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, y, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 50, step=1)) 
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=0.95, color='r', linestyle='-')
plt.text(0.5, 0.85, '95% cut-off threshold', color = 'red', fontsize=16)

ax.grid(axis='x')
plt.show()
```

The chart above tell us that the best number of components to extract from PCA is 45, so let's do it!

```python
neigh_pca_clustering = PCA(n_components=45).fit_transform(neighbourhoods_groupped_clustering)
```

```python
scores_pca = []
for k in np.arange(2,9):
    clusters = KMeans(n_clusters=k, random_state=False).fit(neigh_pca_clustering)
    score = metrics.silhouette_score(neigh_pca_clustering, clusters.labels_)
    scores_pca.append({"k_number":k, "score": score})

pd.DataFrame(scores_pca).plot(x="k_number", y="score", figsize=(12,6))
plt.title("Number of clusters according Silhouette Score (PCA)")
plt.xlabel("K")
plt.ylabel("Score")
```

In this chart we can see that the silhouette score improved a little, then we are gonna use k = 7.

```python
k_number = 4
kmeans_parameters = {
    "n_clusters": [k_number],
    "init": ["k-means++", "random"],
    "algorithm": ["auto", "full", "elkan"],
    "n_init": [2,4,5,6,8,10],
    "random_state": [False]
}
gs_kmeans = GridSearchCV(KMeans(), param_grid=kmeans_parameters)
gs_kmeans.fit(neighbourhoods_groupped_clustering)
```
